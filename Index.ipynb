```python
#索引层的交换
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
L1,L2,L3 = ['A','B'],['a','b'],['alpha','beta']
mul_index1 = pd.MultiIndex.from_product([L1,L2,L3],
    names=('Upper', 'Lower','Extra'))

L4,L5,L6 = ['C','D'],['c','d'],['cat','dog']
mul_index2 = pd.MultiIndex.from_product([L4,L5,L6],names=('Big', 'Small', 'Other'))

df_ex = pd.DataFrame(np.random.randint(-9,10,(8,8)),index=mul_index1,columns=mul_index2)
df_ex
```

```python
df_ex.swaplevel(0,2,axis=1).head()
#列索引的第一和第三行交换(axis=0行索引，axis=1列索引)swaplevel只能交换两个层
```

```python
df_ex.reorder_levels([2,1,0],axis=0).head()
#列表中的数字代表的是原来索引层的位置
```

```python
df_ex.droplevel(1,axis=1).head()
#删除第二层的索引
```

```python
df_ex.droplevel([0,2],axis=0).head()
```

```python
df_ex.rename_axis(index={'Upper':'Changed_row'},columns={'Big':'Changed_col'})
#对索引进行重命名，常用的修改方式是传入字典的映射：
```

```python
df_ex.rename(index={'A':'A1'},columns={'C':'C1'},level=0).head()
#对索引的值进行重命名需要加上层数
```

```python
df_ex.rename(index=lambda x:str.upper(x),level=2).head()
```

```python
#可以对整个索引的值进行替换
new_values=iter(list('abcdefgh'))
df_ex.rename(index=lambda x:next(new_values),level=2)
```

```python
#map函数
df_temp=df_ex.copy()
new_idx = df_temp.index.map(lambda x: (x[0], x[1],str.upper(x[2])))
df_temp.index=new_idx
df_temp
#保留行索引的第一二个元素，将第三个元素转为大写
```

```python
df_temp=df_ex.copy()
new_idx=df_temp.index.map(lambda x:(x[0]+'-'+x[1]+'-'+x[2]))
df_temp.index=new_idx
df_temp
#map函数可以对多级索引进行压缩
```

```python
#反向展开
new_idx=df_temp.index.map(lambda x:tuple(x.split('-')))
df_temp.index=new_idx
df_temp
```

```python
#索引的设置与重置
#索引的设置可以使用set_index 完成，这里的主要参数是append ，表示是否来保留原来的索引，直接把新设定的添加到原索引的内层
df_new=pd.DataFrame({'A':list('abcd'),
                   'B':list('CDFG'),
                   'C':[1,2,3,4],'D':[4,5,6,7]})
df_new.set_index(['A','B'],append=False)
```

```python
df_new=pd.DataFrame({'A':list('abcd'),
                   'B':list('CDFG'),
                   'C':[1,2,3,4],'D':[4,5,6,7]})
df_new.set_index(['A','B'],append=True)
```

```python
df_new.reset_index(drop=False)
```

```python
#reset_index 是set_index 的逆函数，其主要参数是drop ，表示是否要把去掉的索引层丢弃，而不是添加到列中
df_new.reset_index(drop=True)
```

```python
df_set_1=pd.DataFrame([[2,4],[1,3],[4,5],[6,7]],
                      index=pd.Index(['a','b','a','c'],name='id1'))
df_set_2=pd.DataFrame([[1,4],[7,11],[6,7],[13,17]],
                     index=pd.Index(['b','c','c','b'],name='id2'))

# df_set_1 = pd.DataFrame([[0,1],[1,2],[3,4]],
#                         index = pd.Index(['a','b','a'],name='id1'))
# df_set_2 = pd.DataFrame([[4,5],[2,6],[7,1]],
#                         index = pd.Index(['b','b','c'],name='id2'))
df_set_1
```

```python
df_set_2
```

```python
id1,id2 = df_set_1.index.unique(),df_set_2.index.unique()
```

```python
id1.intersection(id2)#取交集
#id1 & id2
```

```python
id1.union(id2)#取并集
#id1 | id2
```

```python
id1.difference(id2)
#(id1 ^ id2) & id1
#第一个集合相对于第二个集合的补集
```

```python
id1.symmetric_difference(id2)
#id1^id2
#仅属于其中一个集合并且不属于两个集合的元素的集合
```
