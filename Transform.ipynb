```python
import pandas as pd
import numpy as np
df=pd.DataFrame({'class':[1,1,2,2],
                 'Name':['san Zhang','san Zhang','si Li','si Li'],
                 'Subject':['Math','English','Math','English'],
                 'Grade':[90,80,70,85]})
df
#pivot 是一种典型的长表变宽表的函数
```

```python
pd.DataFrame({'Gender':['F','F','M','M'],
              'Height':[163, 160, 175, 180]})
```

```python
pd.DataFrame({'Height: F':[163, 160],
              'Height: M':[175, 180]})
```

```python
df.pivot(index=['class','Name'],
         columns='Subject',
         values='Grade')
# 将行索引更改为class,Name，列索引更改为Subject
# 原表中的index 和columns 对应两个列的行组合必须唯一
```

```python
df = pd.DataFrame({'Class':[1, 1, 2, 2, 1, 1, 2, 2],
                   'Name':['San Zhang', 'San Zhang', 'Si Li', 'Si Li','San Zhang', 'San Zhang', 'Si Li', 'Si Li'],
                   'Examination': ['Mid', 'Final', 'Mid', 'Final','Mid', 'Final', 'Mid', 'Final'],
                   'Subject':['Chinese', 'Chinese', 'Chinese', 'Chinese','Math', 'Math', 'Math', 'Math'],
                   'Grade':[80, 75, 85, 65, 90, 85, 92, 88],
                   'rank':[10, 15, 21, 15, 20, 7, 6, 2]})
df
```

```python
df.pivot(index = ['Class','Name'],
         columns = ['Subject','Examination'],
        values = ['Grade','rank'])
```

```python
df = pd.DataFrame({'Name':['San Zhang', 'San Zhang', 'San Zhang', 'San Zhang','Si Li', 'Si Li', 'Si Li', 'Si Li'],
                   'Subject':['Chinese', 'Chinese', 'Math', 'Math','Chinese', 'Chinese', 'Math', 'Math'],
                   'Grade':[80, 90, 100, 90, 70, 80, 85, 95]})
df
```

```python
df.pivot_table(index='Name',columns='Subject',values='Grade',aggfunc=lambda x:x.mean())
```

```python
df.pivot_table(index='Name',columns='Subject',values='Grade',aggfunc='mean',margins=True)
#pivot_table处理重复数据，利用聚合函数aggfunc取均值
#pivot_table 具有边际汇总的功能，可以通过设置margins=True 来实现
```

```python
df = pd.DataFrame({'Class':[1,2],'Name':['San Zhang','Si Li'],'Chinese':[80,75],'Math':[90,80]})
df                                       
```

```python
df.melt(id_vars=['Class','Name'],var_name='Subject',value_vars=['Chinese','Math'],value_name='Grade')
```

```python
df_melted=df.melt(id_vars=['Class','Name'],
                 value_vars=['Chinese','Math'],
                 var_name='Subject',
                 value_name='Grade')
df_melted
```

```python
df_unmelted = df_melted.pivot(index=['Class','Name'],columns='Subject',values='Grade')
df_unmelted
```

```python
df_unmelted = df_melted.pivot(index = ['Class', 'Name'],columns='Subject',values='Grade')
df_unmelted = df_unmelted.reset_index().rename_axis(columns={'Subject':''})
df_unmelted
```

```python
df=pd.DataFrame(np.ones((4,2)),
                index=pd.Index([('A','cat','big'),('A','dog','small'),('B','cat','big'),('B','dog','small')]),
                columns=['col1','col2'])
df
```

```python
df.unstack()
#unstack 的主要参数是移动的层号，默认转化最内层，移动到列索引的最内层，同时支持同时转化多个层
```

```python
df.unstack([0,2])
#将第一层和第三层转化到列索引的最内层
#类似于pivot 中的唯一性要求，在unstack 中必须保证被转为列索引的行索引层和被保留的行索引层构成的组合是唯一的
```

```python
df=pd.DataFrame(np.ones((4,2)),
                index=pd.Index([('A','cat','big'),('A','dog','small'),('B','cat','big'),('B','dog','small')]),
                columns=['col1','col2']).T
df
```

```python
df.stack()
#stack 的作用就是把列索引的层压入行索引，其用法完全类似
```

```python
df.stack([1,2])
#将第二层和第三层转化到行索引的最内层
```
